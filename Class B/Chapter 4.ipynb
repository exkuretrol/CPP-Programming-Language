````c++
int a, b, c, aa;
a = 1;
b = 2;
c = 3;
a = b < c;
aa = b > c;
````

````c++
cout << a;
````

````c++
cout << aa;
````

````c++
int d, e;
d = 1;
e = 1;
if (d = e) cout << d << e;
````

````c++
int x = 5;
cout << x++ << endl;
cout << x;
````

````c++
cout << ++x;
````

````c++
x = 5;
cout << x++;
cout << x;
````

````c++
x = 5;
cout << ++x;
cout << x;
````

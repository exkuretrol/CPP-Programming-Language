```cpp
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;

bool compare(int a, int b)
{
    int ma = a % m;
    int mb = b % m;
    // 負數處理
    int aeo = (a % 2 + 2) % 2, beo = (b % 2 + 2) % 2;
    
    if (ma != mb) return ma < mb;
    if (aeo != beo) return aeo > beo;
    if (aeo) return a > b;
    else return a < b;
}

int main()
{
    int n, m;
    while (cin >> n >> m)
    {
        cout << n << " " << m << endl;
        if (n == 0) break;
        vector<int> a(n);
        for (auto &v : a) cin >> v;
        sort(a.begin(), a.end(), compare);
        for (auto &v : a) cout << v << endl;
    }
}
```

```cpp
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;


class CLESS
{
    int mod;
    CLESS(int m) : mod(m) {}
    bool operator()(int x, int y)
    {
        if (x % mod != y % mod) return x % mod < y % mod;
        // int xeo = ((x % 2) + 2) % 2;
        int xeo = !!(x % 2);
        int yeo = !!(y % 2);
        if (xeo!yeo) return xeo > yeo;
        if (xeo) return x > y;
        else return x < y;
    }
};

int main()
{
    // CLESS cmp(3);
    // cout << cmp(234, 543) << endl;
    // cmp(234, 543) 
    // vvvvvvvvvvvvvvvvvvvvvvv
    // cmp.operator()(234,543);
    
    while (cin >> n >> m)
    {
        CLESS cmp(m);
        cout << n << " " << m << endl;
        if (n == 0) break;
        vector<int> a(n);
        for (auto &v : a) cin >> v;
        sort(a.begin(), a.end(), cmp));
        // sort(a.begin(), a.end(), CLESS(m)));
        for (auto &v : a) cout << v << endl;
    }
    return 0;
}
```

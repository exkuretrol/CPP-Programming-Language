```c++
#include <iostream>
#include <string>
using namespace std;
```

```c++
int n;
while (true)
{
    cin >> n;
    if (n == -1) break;
    cout << n << "的四進位為：";
    int f[] = {0, 0, 0, 0, 0, 0, 0, 0};
    
    for (int i = 0; n > 0; i++)
    {
        f[i] = n % 4;
        n /= 4;
    }
    for (int j = 7; j >= 0; j--)
    {
        cout << f[j];
    }
    cout << endl;
}
```

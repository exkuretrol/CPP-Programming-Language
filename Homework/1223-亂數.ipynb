```c++
#include <cstdlib>
#include <iostream>
using namespace std;
```

```c++
while (true)
{
    int n;
    cin >> n; // 5 9 2 -1
    if (n == -1) break;
    int result[6] = {0, 0, 0, 0, 0, 0};
    for (int i = 0; i < n; i++)
    {
        result[rand() % 6] += 1;
    }
    cout << n << endl;
    for (int i = 0; i < 6; i++)
    {
        printf("%d點:%d次，機率為%.4f%%\n", i + 1, result[i],  (double) result[i] * 100 / n);
    }
}
```

```c++
double avg(int a, int b, int c)
{
    return (a + b + c) / 3;
}
int a, b, c;
cout << "輸入a值:";
cin >> a;
cout << endl << "輸入b值:";
cin >> b;
cout << endl << "輸入c值:";
cin >> c;
cout << "a=" << a << " b=" << b << " c=" << c << endl;
printf("avg(a, b, c)=%.5f", avg(a, b, c));
```

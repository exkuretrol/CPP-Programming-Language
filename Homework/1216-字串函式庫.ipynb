```c++
#include <iostream>
#include <string>
using namespace std;
```

```c++
string s;
while (true)
{
    cout << "輸入一個字串（最多10個）：";
    cin >> s;
    cout << s << endl;
    if (s == "-1") break;
    if (s.size() > 10) break;
    for (int i = 0; i < s.size(); i++) s[i] = islower(s[i]) ? toupper(s[i]) : tolower(s[i]);
    cout << s << endl;
}
```

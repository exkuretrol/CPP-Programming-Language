### 8.1.4 找出陣列的最大最小值

```c++
#include <iostream>
using namespace std;
```

```c++
int n, max = -9999, min = 9999;
while (cin >> n)
{
    if (n == -1) break;
    if (n > max) max = n;
    if (n < min) min = n;
}
cout << max << endl << min;
```

```c++
int A[] = {48, 75, 30, 17, 62};
int i, min = A[0], max = A[0];
int length = sizeof(A) / sizeof(int);
cout << "elements in array A are ";
for (i = 0; i < length; i++)
{
    cout << A[i] << " ";
    if (A[i] > max) max = A[i];
    if (A[i] < min) min = A[i];
}
cout << endl << "Maximum is " << max;
cout << endl << "Minimum is " << min;
```

```c++
#define MAX 5
int score[MAX];
int i = 0, num;
float sum = 0.0f;
cout << "Enter 0 stopping input!!" << endl;
do
{
    if (i == MAX)
    {
        cout << "No more space!!" << endl;
        i++;
        break;
    }
    cout << "Input score: ";
    cin >> score[i];
} while (score[i++] > 0);
num = i - 1;
for (i = 0; i < num; i++) sum += score[i];
cout << "Average of all is " << sum / num << endl;
```
